$$E[y] = \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot (n \cdot p_1) \cdot (n \cdot p_2) \cdot (n \cdot p_3) ...  (n \cdot p_k)$$

$$E[y] = n \cdot p_i$$

Consider the following experiment:

1. The experiment consists of a fixed number, $n$, of identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to $q = (1-p)$.
4. The trials are independent.
5. The random variable of interest is $Y$, the number of successes observed during $n$ trials.

Question 1: From the following steps shown above, derive the probability distribution of the experiment.

Question 2: Show that the expectation of this probability distribution is $E[y] = \Sigma_y \cdot P(y) = np$.

Question 3: Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Question 4: Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective.

**Solution:**

### Question 1:
From the following steps shown above, derive the probability distribution of the experiment.

- Define the random variable:

Let $Y$ be the random variable representing the number of successes in $n$ trials. Each trial results in a success (S) with probability $p$ and a failure (F) with probability $q = 1 - p$.

- Consider a specific outcome:

Suppose we have a specific outcome sequeance of the $n$ trials that results in exactly $y$ successes and $n - y$ failures.

The probability of observing one specific sequence with exactly $y$ successes and $n - y$ failures is given by the product of the probabilities of each individual trial outcome:
$$P(\text{specific sequence with} \ y \ \text{successes}) = p^y \cdot q^{n-y}$$

- Count the number of such sequences

$$\begin{pmatrix}
n \\
y
\end{pmatrix}
= \frac{n!}{y!(n - y)!}$$

The number of different sequences that result in exactly $y$ successes out of $n$ trials is represented by the binomial coefficient above. It represents the number of ways to choose $y$ successes from $n$ trials.

- Combine both probability and count

$$P(Y = y) =
\begin{pmatrix}
n \\
y
\end{pmatrix}
p^y q^{n-y}$$

- The probability distribution of the experiment is:

$$P(Y = y) =
\begin{pmatrix}
n \\
y
\end{pmatrix}
p^y q^{n-y} \\

\text{or} \\

P(Y = y) =
\begin{pmatrix}
n \\
y
\end{pmatrix}
p^y (1 - p)^{n-y}$$

### Question 2

Show that the expectation of this probability distribution is $E[y] = \Sigma_y \cdot P(y) = np$.

- Expected Value calculation

$$E[Y] = \sum_{y=0}^n y \cdot P(Y = y)$$

Substitute the probability mass function of the binomial distribution.

$$E[Y] = \sum_{y=0}^n y \cdot 
\begin{pmatrix}
n \\
y
\end{pmatrix} 
p^y (1 - p)^{n-y}$$

- Adjust the Summation:

$y \cdot \begin{pmatrix} n \\ y \end{pmatrix}$ can be rewritten using the property of binomial coefficients.

$$y \cdot \begin{pmatrix} n \\ y \end{pmatrix} = n \cdot \begin{pmatrix} n - 1 \\ y - 1 \end{pmatrix}$$

Substitute this into the summation.

$$E[Y] = \sum_{y=1}^n n \cdot 
\begin{pmatrix}
n - 1 \\
y - 1
\end{pmatrix} 
p^y (1 - p)^{n-y}$$

(Note that we changed the summation to start from $y = 1$ because $y = 0$ does nothing to the expectation.)

- Factor out $n$

$$E[Y] = n \cdot \sum_{y=1}^n 
\begin{pmatrix}
n - 1 \\
y - 1
\end{pmatrix} 
p^y (1 - p)^{n-y}$$

- Change the index of summation

Let $k = y - 1$. As $y$ goes from 1 to $n$, $k$ goes from 0 to $n - 1$

$$E[Y] = n \cdot \sum_{k=0}^{n-1} 
\begin{pmatrix}
n - 1 \\
k
\end{pmatrix} 
p^{k+1} (1 - p)^{(n-1)-k}$$

- Separate $p$ from the sum

$$E[Y] = n \cdot p \cdot 
\begin{pmatrix}
n - 1 \\
k
\end{pmatrix} 
p^k (1 - p)^{(n-1)-k}$$

- Recognize Binomial Expansion

The summation now resembles the binomial expansion of $(p + (1 - p))^{n-1}$:

$$E[Y] = n \cdot p \cdot 
\begin{pmatrix}
n - 1 \\
k
\end{pmatrix} 
p^k (1 - p)^{(n-1)-k} = (p + (1 - p))^{n-1} = 1$$

Therefore,

$$E[Y] = n \cdot p \cdot 1 = np$$

### Question 3

Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover?

- We are given:

    $n$ = 10 \
    Probability of recovery = 30% or 0.30 \
    Supposed number of recovery = 9

Find: Probability that at least 9 out of 10 patients receiving the medication will recover.

- Use the Binomial Probability Mass Function

$$P(Y \geq 9) = P(Y = 9) + P(Y = 10)$$
$$P(Y \geq 9) = \binom{10}{9} \cdot (0.30)^9 \cdot (1 - 0.30)^{10-9} + \binom{10}{10} \cdot (1 - 0.30)^{10-10}$$
$$P(Y \geq 9) = 10 \cdot (0.30)^9 \cdot (0.70)^1 + 1 \cdot (0.30)^{10}  \cdot (0.70)^0$$
$$P(Y \geq 9) = 10 \cdot (0.30)^9 \cdot 0.70 + (0.30)^{10}$$
$$P(Y \geq 9) = 0.000137781 + 0.0000059049$$
$$P(Y \geq 9) \approx 0.0001436869 \text{or} 0.014 \%$$

The probability of at least 9 out of 10 patients receiving the medication that will recover is 0.014%.

Consider the probability distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$

Question 1: Find a general formula for the expected value of this distribution.

Question 2: A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. **Solution:**

### Question 1

Find a general formula for the expected value of this distribution.

The given probability distribution is the probability mass function of a Poisson distribution with parameter $\lambda$. 

- Expected value calculation

$$E[Y] = \sum_{y=0}^\infty y \cdot P(Y = y)$$

Substitute $P(Y = y) = \frac{\lambda^y}{y!} e^{-\lambda}$, where $y = 0, 1, 2, ...$

$$E[Y] = \sum_{y=0}^\infty y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$$

- Simplify:

$$E[Y] = \sum_{y=1}^\infty y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$$

(The term $y = 0$ does nothing to the expectation, so we start at $y = 1$.)

Notice that $y \cdot \frac{\lambda^y}{y!} = \lambda \cdot \frac{\lambda^{y-1}}{(y-1)!}$.

$$E[Y] = \lambda e^{-\lambda} \sum_{y=1}^\infty \frac{\lambda^{y-1}}{(y - 1)!}$$

Let $k = y - 1$. As $y$ goes from 1 to $\infty$, $k$ goes from 0 to $\infty$.

$$E[Y] = \lambda e^{-\lambda} \sum_{k=0}^\infty \frac{\lambda^{k}}{k!}$$

The summation is the series expansion of the exponential function $e^\lambda$.

$$\sum_{k=0}^\infty \frac{\lambda^{k}}{k!} = e^\lambda$$

Thus:

$$E[Y] = \lambda e^{-\lambda} \cdot e^\lambda = \lambda$$
$$E[Y] = \lambda$$
